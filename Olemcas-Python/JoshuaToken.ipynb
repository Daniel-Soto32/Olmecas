```python
import tkinter as tk
import time


class Interfaz:
    def __init__(self, ventana):
        #Inicializar la ventana con un título
        self.ventana=ventana
        self.ventana.title("Wazuh API")
        self.ventana.resizable(True,True)


        #Drop-down vulnerabilidades
        grado_vulnerabilidad=tk.StringVar(ventana)
        grado_vulnerabilidad.set("ANY")
        lista_riesgo=["ANY","LOW", "MID", "HIGH",]
        self.opcion_riesgo=tk.OptionMenu(ventana,grado_vulnerabilidad, *lista_riesgo)
        self.opcion_riesgo.grid(row=0, column=1, columnspan=1,  padx=5, pady=5)

        #Label agente-vulnerabilidades
        self.labelV=tk.Label(ventana, text="Show agents with _____ vulnerabilities")
        self.labelV.grid(row=0, column=0, columnspan=1,  padx=5, pady=5, sticky="W")

        #Agrega una caja de texto para que sea la pantalla de impresión
        self.pantalla=tk.Text(ventana, state="disabled", width=60, height=6, background="#F0F0FF", foreground="black", font=("Arial",16), yscrollcommand=True)
        self.pantalla.grid(row=4, column=0, columnspan=8,  padx=5, pady=5)

        
        #Función botón obtener vulnerabilidades
        def get_Vulnerabilities():
            print("Imprimiendo...")
            self.pantalla.configure(state="normal")
            self.pantalla.insert(tk.END,"Imprimiendo...")
            self.pantalla.configure(state="disabled")

        #Boton Obtener Vulnerabilidades
        self.btn_obtener =tk.Button(ventana, text="Obtener Vulnerabilidades", command=get_Vulnerabilities)
        self.btn_obtener.grid(row=0, column=3, columnspan=1,  padx=5, pady=5)

        #Label buscador vulnerabilidades
        self.labelbusc_vul=tk.Label(ventana, text="Search vulnerability: ")
        self.labelbusc_vul.grid(row=1, column=0, columnspan=1,  padx=5, pady=5, sticky="W")
        
        #Buscador vulnerabilidades
        self.buscador_vulnerabilidades=tk.Entry(ventana, width=32, background="#F0F0FF", foreground="black", font=("Arial",10) )
        self.buscador_vulnerabilidades.grid(row=1, column=0, columnspan=2,  padx=5, pady=5, sticky="E")

        
        #Drop-down opciones de agentes
        accion_agente=tk.StringVar(ventana)
        accion_agente.set("UPDATE")
        lista_agente=["UPDATE","RESTART","DELETE"]
        self.opcion_agente=tk.OptionMenu(ventana,accion_agente, *lista_agente)
        self.opcion_agente.grid(row=2, column=0, columnspan=1,  padx=5, pady=5)

        #Label agente-vulnerabilidades
        str_out=tk.StringVar(ventana)
        str_out.set("")
        self.labelA=tk.Label(ventana, textvariable=str_out)

        #Función botón acción sobre agente
        def activty_Agent(param_accion_agente):
            self.pantalla.configure(state="normal")

            if param_accion_agente == "UPDATE":
                self.pantalla.insert(tk.END,"Updating Agent...")
                print("Updating Agent...")
            elif param_accion_agente =="RESTART":
                self.pantalla.insert(tk.END,"Restarting Agent...")
                print("Restarting Agent...")
            elif param_accion_agente =="DELETE":
                self.pantalla.insert(tk.END,"Deleting Agent...")
                print("Deleting Agent...")
            else:
                print("No hay valor")

            self.pantalla.configure(state="disabled")

        #Boton acción sobre agente
        self.btn_agent =tk.Button(ventana, text="Execute", command=lambda:activty_Agent(accion_agente.get()))
        self.btn_agent.grid(row=2, column=1, columnspan=1,  padx=5, pady=5)

        #Actualizar opcion de riesgo
        def actualizar_riesgo(*args):
             gv=grado_vulnerabilidad.get()
             self.labelV.config(text="Show agents with "+ gv +" vulnerabilities")

        grado_vulnerabilidad.trace_add("write", actualizar_riesgo)


        #Actualizar opcion de agente
        def actualizar_opcion_agente(*args):
             str_out.set(accion_agente.get())
             print(str_out.get())

        accion_agente.trace_add("write", actualizar_opcion_agente)

ventana_principal=tk.Tk()
ventana_principal.columnconfigure(8)
interfaz_API=Interfaz(ventana_principal)
ventana_principal.mainloop()

```
