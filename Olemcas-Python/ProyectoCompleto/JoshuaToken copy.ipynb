```python
from main import *

class Interfaz:
  def __init__(self, ventana):
     #Inicializar la ventana con un título
     self.ventana = ventana
     self.ventana.title("Wazuh API")
     self.ventana.iconbitmap('assets/wazuh.ico')
     self.ventana.resizable(True,True)
     self.ventana.attributes("-fullscreen",True)
     self.ventana.configure(bg="skyblue")
     menuBar(ventana)

     #Tamaño y posición de ventana
     anchura_ventana = ventana.winfo_reqheight()
     altura_ventana = ventana.winfo_reqwidth()
     anchura_monitor = ventana.winfo_screenwidth()
     altura_monitor = ventana.winfo_screenheight()
     coordinadaX = (anchura_monitor/2)-(anchura_ventana/2)
     coordinadaY = (altura_monitor/2)-(altura_ventana/2)
     self.ventana.geometry(f'{anchura_ventana}x{altura_ventana}+{int(coordinadaX)}+{int(coordinadaY)}')

     #Estilos y tamaño de letra
     fontStyle_Labels = tkFont.Font( family = "Lucida Grande", size = 20)
     fontStyle_LittleLabels = tkFont.Font( family = "Lucida Grande", size = 14)
     fontStyle_Text = tkFont.Font( family = "Helvetica bold", size = 30)

     #Label Sección Vulnerabilidades
     self.labelVulnerabilidades = tk.Label(ventana, text="Vulnerabilities", bg="skyblue", font = fontStyle_Labels, anchor = "center")
     self.labelVulnerabilidades.grid(row=0, column=0, columnspan=4,  padx=5, pady=5, sticky = "W")

     self.labelInfo = tk.Label(ventana, text="Server Status and Info", bg="skyblue", font = fontStyle_Labels)
     self.labelInfo.grid(row=0, column=10, columnspan=4,  padx=5, pady=5, sticky = "W")

     #Label agente-vulnerabilidades
     self.labelV = tk.Label(ventana, text="SHOW _____ RISK VULNERABILITIES")
     self.labelV.grid(row=1, column=0, columnspan=1,  padx=5, pady=5, sticky="W")

     self.labelI = tk.Label(ventana, text="SHOW INFO ABOUT: ")
     self.labelI.grid(row=3, column=10, columnspan=1,  padx=5, pady=5, sticky="W")

     #Drop-down vulnerabilidades
     grado_vulnerabilidad = tk.StringVar(ventana)
     grado_vulnerabilidad.set("ANY")
     lista_riesgo = ["CRITICAL","LOW", "MID", "HIGH",]
     self.opcion_riesgo = tk.OptionMenu( ventana,grado_vulnerabilidad, *lista_riesgo)
     self.opcion_riesgo.grid( row=1, column=1, columnspan=1,  padx=5, pady=5, sticky="W")
        
     # PANTALLA 1
     self.pantalla1=tk.Text(ventana, state="disabled", width=38, height=25, background="#D1E9F1", foreground="black", font=("Arial",16), yscrollcommand=True)
     self.pantalla1.grid(row=5, column=0, columnspan=4,  padx=5, pady=5, sticky="W")

     #Boton Obtener Vulnerabilidades
     self.btn_obtener_vulnerabilidades = tk.Button(
         ventana, 
         text="GET VULNERABILITES", 
         command = lambda:get_Vulnerabilities( 
             self.pantalla1, 
             grado_vulnerabilidad.get()
             )
         )
     self.btn_obtener_vulnerabilidades.grid(row=1, column=1, columnspan=1,  padx=5, pady=5, sticky="E")

     #Label buscador vulnerabilidades
     self.labelbusc_vul=tk.Label(ventana, text="SEARCH VULNERABILITY: ")
     self.labelbusc_vul.grid(row=2, column=0, columnspan=1,  padx=5, pady=5, sticky="W")
        
     #Buscador vulnerabilidades
     self.buscador_vulnerabilidades= tk.Entry(ventana, width=32, background="#F0F0FF", foreground="black", font=("Arial",10) )
     self.buscador_vulnerabilidades.grid(row=2, column=1, columnspan=1,  padx=5, pady=5)

     #Boton buscar Vulnerabilidades
     self.btn_buscar_vulnerabilidades =tk.Button(
          ventana, 
          text=u"\u26b2", 
          command = lambda:search_Vulnerabilities(self.pantalla1),
          )
     self.btn_buscar_vulnerabilidades.grid(row=2, column=2, columnspan=1,  padx=5, pady=5, sticky="W")

     #Boton TOP 10 vulnerabilidades
     self.btn_T10vulnerabilites = tk.Button(
          ventana, 
          text="TOP 10 VULNERABILITES", 
          command = lambda:get_T10Vulnerabilities(self.pantalla1),
          )
     self.btn_T10vulnerabilites.grid(row=3, column=0, columnspan=1,  padx=5, pady=5, sticky="W")

     #Label Sección Agentes
     self.labelAgentes=tk.Label(ventana, text="Agents", bg="skyblue", font=fontStyle_Labels, anchor="center").grid(row=0, column=5, columnspan=4,  padx=5, pady=5, sticky="W")

     # PANTALLA 2
     self.pantalla2 = tk.Text(ventana, state="disabled", width=38, height=25, background="#D1E9F1", foreground="black", font=("Arial",16), yscrollcommand=True)
     self.pantalla2.grid(row=5, column=5, columnspan=4,  padx=5, pady=5, sticky="W")
        
     #Botón lista agentes
     self.btn_agent_list =tk.Button(
          ventana, 
          text="GET AGENT LIST", 
          command = lambda:get_AgentList(self.pantalla2),
          ).grid(row=1, column=5, columnspan=1,  padx=5, pady=5, sticky="W")

     #Boton TOP 10 agentes vulnerables
     self.btn_T10agents =tk.Button(
         ventana, 
         text="TOP 10 VULNERABLE AGENTS", 
         command = lambda:get_T10Agents(self.pantalla2),
         ).grid(row=1, column=6, columnspan=1,  padx=5, pady=5, sticky="W")

     #Drop-down opciones de agentes
     accion_agente=tk.StringVar(ventana)
     accion_agente.set("UPDATE")
     lista_agente=["UPDATE","RESTART","DELETE"]
     self.opcion_agente=tk.OptionMenu(ventana,accion_agente, *lista_agente)
     self.opcion_agente.grid(row=2, column=5, columnspan=1,  padx=5, pady=5, sticky="W")

     #Label agente-vulnerabilidades
     str_out=tk.StringVar(ventana)
     str_out.set("")
     self.labelA=tk.Label(ventana, textvariable=str_out)

     #Botón acción sobre agente
     self.btn_agent = tk.Button(
          ventana, 
          text="EXECUTE", 
          command = lambda:activty_Agent( 
             self.pantalla2, 
             accion_agente.get() 
             )
          )
     self.btn_agent.grid(row=2, column=6, columnspan=1,  padx=5, pady=5)

     #Actualizar opcion de riesgo
     def actualizar_riesgo(*args):
             gv = grado_vulnerabilidad.get()
             self.labelV.config(text="SHOW "+ gv +" RISK VULNERABILITIES")
     grado_vulnerabilidad.trace_add("write", actualizar_riesgo)

     #Actualizar opcion de agente
     def actualizar_opcion_agente(*args):
             str_out.set(accion_agente.get())
             print(str_out.get())
     accion_agente.trace_add("write", actualizar_opcion_agente)

     # PANTALLA 3
     self.pantalla3 = tk.Text(ventana, state="disabled", width=38, height=25, background="#D1E9F1", foreground="black", font=("Arial",16), yscrollcommand=True)
     self.pantalla3.grid(row=5, column=10, columnspan=4,  padx=5, pady=5, sticky="W")

     #Información de inventario 
     grado_info = tk.StringVar(ventana)
     grado_info.set("CHOOSE ONE")
     lista_info = ["HARDWARE","HOTFIXES", "NETADDR", "NETIFACE","NETPROTO","OS","PACKAGES","PORTS","PROCESSES"]
     self.opcion_info = tk.OptionMenu( ventana, grado_info, *lista_info)
     self.opcion_info.grid(row=1, column=10, columnspan=1,  padx=5, pady=5, sticky="W")

     self.btn_obtener_info =tk.Button(
          ventana, 
          text="GET INFO", 
          command=lambda:get_Info(
             self.pantalla3,
             grado_info.get()
             )
         ).grid(row=1, column=11, columnspan=1,  padx=5, pady=5, sticky="E")
        
     #Info de estado de servidor Wazuh
     grado_server = tk.StringVar(ventana)
     grado_server.set("STATUS SERVER")
     lista_server = ["HARDWARE","HOTFIXES", "NETADDR", "NETIFACE","NETPROTO","OS","PACKAGES","PORTS","PROCESSES"]
     self.opcion_serer = tk.OptionMenu( ventana, grado_server, *lista_server)
     self.opcion_serer.grid(row=2, column=10, columnspan=1,  padx=5, pady=5, sticky="W")

     self.btn_obtener_server =tk.Button(
          ventana, 
          text="GET STATUS SERVER", 
          command=lambda:get_Info(
             self.pantalla3,
             grado_server.get()
             )
         ).grid(row=2, column=11, columnspan=1,  padx=5, pady=5, sticky="E")

     #Scrolls 
     scrolls(self.pantalla1,4)
     scrolls(self.pantalla2,9)
     scrolls(self.pantalla3,14)
       
ventana_principal=tk.Tk()
#ventana_principal.columnconfigure(8)
interfaz_API=Interfaz(ventana_principal)
ventana_principal.mainloop()

```
